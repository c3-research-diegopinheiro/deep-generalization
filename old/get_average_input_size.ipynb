```python
import os
import seaborn as sns
from matplotlib.image import imread
import numpy as np
from dataset_path import dataset_path

data_dir = dataset_path
test_path = data_dir+'/test'
train_path = data_dir+'/train'
```

```python
dim1 = []
dim2 = []

for image_filename in os.listdir(test_path+'/no'):
    img = imread(test_path + '/no/' + image_filename)
    d = img.shape
    dim1.append(d[0])
    dim2.append(d[1])

for image_filename in os.listdir(test_path+'/yes'):
    img = imread(test_path + '/yes/' + image_filename)
    d = img.shape
    dim1.append(d[0])
    dim2.append(d[1])

for image_filename in os.listdir(train_path+'/no'):
    img = imread(train_path + '/no/' + image_filename)
    d = img.shape
    dim1.append(d[0])
    dim2.append(d[1])


for image_filename in os.listdir(train_path+'/yes'):
    img = imread(train_path + '/yes/' + image_filename)
    d = img.shape
    dim1.append(d[0])
    dim2.append(d[1])
```

```python
sns.jointplot(dim1, dim2)
```
